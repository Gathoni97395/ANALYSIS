## data cleaning

```python
## handling misisng values
missing_val=sales_df.isnull().sum()
```

```python
print(missing_val)
```

```python
sales_df.dropna(subset=['rating_count'],axis=0, inplace=True)
```

```python
sales_miss=sales_df.isnull().sum()
print(sales_miss)
```

```python
## missing values are handled
```

### data cleaning

```python
## converting the prices colums to numerical format
```

```python
sales_df['discounted_price'] = sales_df['discounted_price'].replace({'₹': '', ',': ''}, regex=True).astype(float)
sales_df['actual_price'] = sales_df['actual_price'].replace({'₹': '', ',': ''}, regex=True).astype(float)
```

```python
print(sales_df.dtypes)
```

```python
sales_df.head(3)
```

```python

#converting rating count column to integer
#sales_df['rating_count'] = sales_df['rating_count'].astype(int)
sales_df['rating_count'] = pd.to_numeric(sales_df['rating_count'], errors='coerce').fillna(0).astype(int)
```

```python
sales_df['discount_percentage'] = sales_df['discount_percentage'].replace({ '%': ''}, regex=True).astype(int)
```

## data transformation and analysis

### data transformation

```python

# calculate the discount amount
sales_df['discount_amount']=sales_df['actual_price']-sales_df['discounted_price']
```

```python
#extract the product name from category
sales_df['main_category'] =sales_df['category'].apply(lambda x: x.split('|')[-1])
```

### data analysis

```python
#checking the distribution  of ratings

plt.figure(figsize=(10, 6))
plt.hist(sales_df['rating'], edgecolor='blue',rwidth=0.8,bins=20,)  
plt.title('Distribution of Ratings')
plt.xlabel('Ratings')
plt.ylabel('Frequency')
plt.show()
```

# exploring the correations among diferrent variables
corr = sales_df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm")

```python
#top rated products
#calculating the rating for each category and grouping them
top_categories = sales_df.groupby('main_category')['rating'].count().sort_values(ascending=False).head(10).reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=top_categories, x='main_category', y='rating') 
plt.title('Top Categories by Rating Counts')
plt.xlabel('Categories')
plt.ylabel('Rating Counts') 
plt.xticks(rotation=45, ha='right')


plt.show()
```

```python
#rename the main category
#creating a pie chart for category  percentage
sales_df = sales_df.rename(columns={'main_category': 'product'})
top_categories=sales_df.groupby('product')['rating'].count().sort_values(ascending=False).head(10)

  # Optional: Adjust figure size
plt.figure(figsize=(8, 8))
top_categories.plot(kind='pie', autopct='%1.1f%%', startangle=140) 
plt.title('Top Products by Rating Counts')
plt.ylabel('') 
plt.show()
```

```python
quantity_sold = sales_df.groupby('product').size().reset_index(name='total_items')
top_10_categories = quantity_sold.sort_values(by='total_items', ascending=False)
print(top_10_categories)
```

```python
#calculating total sales
sales_df['product_quantity'] = sales_df.groupby('product_id')['product_id'].transform('size')
sales_df['total_sales'] = sales_df['discounted_price'] * sales_df['product_quantity']
sales_df.head()
product_sales = sales_df.groupby('product_name')['total_sales'].sum().sort_values(ascending=False).head(10)
print(product_sales)


      



```

```python
# calculating the top categories by sales
category_sales = sales_df.groupby('product')['total_sales'].sum().sort_values(ascending=False).head(10)
print(category_sales)
```

### data visualization


```python
# plotting bar graphs for sales
```

```python
category_sales_sorted = category_sales.sort_values(ascending=False) 

plt.figure(figsize=(10, 6))
bars = sns.barplot(x=category_sales_sorted.values, y=category_sales_sorted.index, palette='viridis')
for bar in bars.patches:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2,
             f'{bar.get_width():,.0f}',
             va='center', ha='left')

plt.title('Top Categories by Total Sales', fontsize=16)
plt.xlabel('Total Sales ($)', fontsize=12)
plt.ylabel('Product Category', fontsize=12)
sns.despine(left=True, bottom=True) 
plt.grid(axis='x', linestyle='--', alpha=0.6) 
plt.tight_layout()
plt.show()
```
